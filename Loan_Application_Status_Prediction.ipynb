## Removing Outliers

## 1. Zscore method using Scipy

```python
#  Outliers will be removed only from Continuous column variable i.e; 'ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term' and 'Credit_History'. 
#We will not remove outliers from Categorical column i.e;  'Gender', 'Dependents', 'Education' and 'Self_Employed'.

variables = df[['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term', 'Credit_History']]

z=np.abs(zscore(variables))

# Creating new dataframe
df_appl = df[(z<3).all(axis=1)]
df_appl.head()
```

```python
z.head()
```

```python
print("Old DataFrame data in Rows and Column:",df.shape)
print("New DataFrame data in Rows and Column:",df_appl.shape)
print("Total Dropped rows:",df.shape[0]-df_appl.shape[0])
```

### Percentage Data Loss using Zscore

```python
loss_percent=(614-577)/614*100
print(loss_percent,"%")
```

## 2. IQR (Inter Quantile Range) method

```python
#1st quantile
Q1=variables.quantile(0.25)

# 3rd quantile
Q3=variables.quantile(0.75)

#IQR
IQR=Q3 - Q1
df_stat=df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
print("Old DataFrame data in Rows and Column:",df.shape)
print("\nNew DataFrame data in Rows and Column:",df_stat.shape)
print("\nTotal Dropped rows:",df.shape[0]-df_stat.shape[0])
```

### Percentage Data Loss using IQR

We can check by using IQR method there is large data loss in comparision to Zscore method. So, we will consider Zscore method.

## Creating Model

#### Finding the best random state among all the models

```python
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.30, random_state =i)
    LR = LogisticRegression()
    LR.fit(x_train, y_train)
    pred = LR.predict(x_test)
    acc=accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu," on Random_state ",maxRS)
```

__At random state 10, we are getting best accuracy score i.e., 78%__

## Creating train-test-split

```python
# creating new train test split using the random state.
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=maxRS)
```
