```python
# HYPER-PARAMETER TUNING
model_params  = {
    "svm":{
        "model":SVC(gamma="auto"),
        "params":{
            'C' : [1,10,20],
            'kernel':["rbf"]
        }
    },
    "decision_tree":{
        "model": DecisionTreeClassifier(),
        "params":{
            'criterion':["entropy","gini"],
            "max_depth":[5,8,9]
        }
    },
    "random_forest":{
        "model": RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10],
            "max_depth":[5,8,9]
        }
    },
    'logistic_regression' : {
        'model' : LogisticRegression(solver='liblinear',multi_class = 'auto'),
        'params': {
            "C" : [1,5,10]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
score=[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp["model"],mp["params"],cv=8,return_train_score=False)
    clf.fit(x,y)
    score.append({
        "Model" : model_name,
        "Best_Score": clf.best_score_,
        "Best_Params": clf.best_params_
    })
```

```python
glass_df=pd.DataFrame(score,columns=["Model","Best_Score","Best_Params"])
glass_df
```

```python
#  XGB CLASSIFIER
def get_best_param_for_xgboost(x_train, y_train):
    xgb = XGBClassifier(objective='binary:logistic')
    param_grid_xgboost = {
        'learning_rate' :[0.5,0.1,0.01,0.001],
        'max_depth':[3,5,10,20],
        'n_estimators':[10,50,100,200]
    }
    grid = GridSearchCV(estimator=xgb, param_grid = param_grid_xgboost, cv = 5 , verbose = 0)
    grid.fit(x_train , y_train)
    learning_rate = grid.best_params_['learning_rate']
    n_estimators = grid.best_params_['n_estimators']
    max_depth = grid.best_params_['max_depth']
    print(learning_rate)
    print(n_estimators)
    print(max_depth)
```
